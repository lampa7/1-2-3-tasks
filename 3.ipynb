```python
def data_prep(data):
    X = data['6']
    y = data['target'] 

    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)

    # NORMALIZATION for distributing all data in range [-1, 1]
    X_train_norm = (X_train - np.mean(X_train)) / (max(X_train) - min(X_train))
    X_test_norm = (X_test - np.mean(X_test)) / (max(X_test) - min(X_test))

    # adding extra dim
    X_train_norm = X_train.values.reshape(-1,1)
    X_test_norm = X_test.values.reshape(-1,1)
    
    return X_train_norm, X_test_norm, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = data_prep(data)
```

```python
def model():
    model = keras.Sequential()
    model.add(keras.layers.Dense(32, input_dim=(X_train.shape[1]), activation='relu'))
    model.add(keras.layers.Dense(1, activation='linear'))
    
    def metrics_rmse(preds, y_test):
        return K.sqrt(K.mean(K.square(preds - y_test)))
        
    model.compile(
        optimizer='adam',
        loss=metrics_rmse,
        metrics=metrics_rmse) 
    return model
```

```python
model = model()
```

```python
history = model.fit(X_train, y_train,
                    validation_data=(X_test, y_test),
                    batch_size=80, epochs=70)
```

```python
def plot_history(history):
    a = [i for i in history.history.keys()]
    plt.plot(history.history[a[1]]) 
    plt.plot(history.history[a[3]]) 
    plt.title('model rmse') 
    plt.ylabel('rmse')
    plt.xlabel('epoch') 
    plt.legend(['train', 'test'], loc='upper left') 
    plt.show()

plot_history(history)
```
